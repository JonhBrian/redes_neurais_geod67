```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from keras import backend
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X,y = make_moons(n_samples=1000, noise=0.15)
```

```python
df = pd.DataFrame(X,columns=['x1','x2'])
df['target'] = y
df.head()
```

```python
fig, axes = plt.subplots()
for i,j in zip(df['target'].unique(),['blue','red']):
    axes.scatter(df[df['target']==i]['x1'],df[df['target']==i]['x2'],c=j,edgecolor='black',alpha=0.7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
plt.scatter(X_test.T[0],X_test.T[1],c=y_test,cmap="bwr",edgecolor='black',alpha=0.7)
plt.tight_layout()
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

## 1 neurnio

```python
model = Sequential()
```

```python
model.add(Dense(1,input_shape=(2,),activation='sigmoid'))
```

```python
model.compile(SGD(learning_rate=0.5),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=300)
```

```python
fig, axes = plt.subplots()
x = np.linspace(df['x1'].min(),df['x1'].max(),200)
y = np.linspace(df['x2'].min(),df['x2'].max(),200)
X,Y = np.meshgrid(x,y)
Z = model.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
axes.contourf(X,Y,Z,cmap='bwr',alpha=0.7)
plt.scatter(df['x1'],df['x2'],c=df['target'],cmap="bwr",edgecolor='black',alpha=0.7)
plt.tight_layout()
```

## Adicionando mais neurnios

```python
backend.clear_session() #Clear para anlise anterior no afetar o novo modelo
```

```python
model = Sequential()
model.add(Dense(4,input_shape=(2,),activation='sigmoid')) #Neurnios
model.add(Dense(1,activation='sigmoid')) #Sada
model.compile(SGD(learning_rate=0.7),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
params = model.fit(X_train,y_train,epochs=500,verbose=0)
```

```python
plt.plot(params.epoch,params.history['loss'])
plt.tight_layout()
```

```python
plt.plot(params.epoch,params.history['accuracy'])
plt.tight_layout()
```

```python
fig,axes = plt.subplots()
axes.plot(np.arange(1,501),params.history['loss'],c='blue')
ax = axes.twinx()
ax.plot(np.arange(1,501),params.history['accuracy'],c='red')
plt.tight_layout()
```

```python
fig, axes = plt.subplots()
x = np.linspace(df['x1'].min(),df['x1'].max(),200)
y = np.linspace(df['x2'].min(),df['x2'].max(),200)
X,Y = np.meshgrid(x,y)
Z = model.predict(np.c_[X.ravel(),Y.ravel()])
Z = Z.reshape(X.shape)
axes.contourf(X,Y,Z,cmap='bwr',alpha=0.7)
plt.scatter(df['x1'],df['x2'],c=df['target'],cmap="bwr",edgecolor='black',alpha=0.7)
plt.tight_layout()
```

```python
model.evaluate(X_test,y_test)
```
