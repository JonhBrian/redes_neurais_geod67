```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('facies_dataset_reg_pe.csv')
df.head()
```

```python
df.columns
```

```python
df.Facies.unique()
```

```python
cols = ['GR','ILD_log10','DeltaPHI','PHIND','NM_M','PHID',
        'PHIN','Formation_numsk','Pe_reg']

X = df[cols]
y = df['Facies']-1 #Devido ao keras que deve começar com 0 nas labels
```

```python
from keras.utils.np_utils import to_categorical
```

```python
y_cat = to_categorical(y)
```

```python
y_cat.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.25, random_state=42)
```

### Construindo uma rede neural (1 camada)

```python
from keras.models import Sequential
from keras.layers import Dense #Garante que todos os neurônios input/output estarão conectados
from keras.optimizers import Adam 
from keras import backend #Limpar layers ao realizar o retreino das redes
```

$$ Softmax $$
$$ g(z) = \frac{e^{z_i}}{\sum^{k}_{j=1}{e^{z_j}}} $$

```python
model = Sequential()
model.add(Dense(9,input_shape=(9,),activation='softmax'))
model.compile(Adam(lr=0.01),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist = model.fit(X_train,y_train,epochs=500,verbose=0)
```

```python
plt.plot(hist.history['loss'])
```

```python
plt.plot(hist.history['accuracy'])
```

### Aplicando standardization aos dados

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaler, y_cat, test_size=0.25, random_state=42)
```

```python
backend.clear_session()
model = Sequential()
model.add(Dense(9,input_shape=(9,),activation='softmax'))
model.compile(Adam(lr=0.01),loss='categorical_crossentropy',metrics=['accuracy']) #Função custo para cada output e realizamos um somatório
```

```python
hist = model.fit(X_train,y_train,epochs=500,verbose=0)
```

```python
plt.plot(hist.history['loss'])
```

```python
plt.plot(hist.history['accuracy'])
```

### Variando a taxa de aprendizado

```python
lr = [0.005, 0.0075, 0.01, 0.03, 0.05, 0.1]
list_models = []
hist_models = []
for param in lr:
    backend.clear_session()
    model = Sequential()
    model.add(Dense(9,input_shape=(9,),activation='softmax'))
    model.compile(Adam(lr=param),loss='categorical_crossentropy',metrics=['accuracy']) 
    hist_models.append(model.fit(X_train,y_train,epochs=500,verbose=0))
    list_models.append(model)
```

```python
for hist,param in zip(hist_models,lr):
    plt.plot(hist.history['loss'],label=str(param))
plt.legend()
plt.show()
```

```python
for hist,param in zip(hist_models,lr):
    plt.plot(hist.history['accuracy'],label=str(param))
plt.legend()
plt.show()
```
